```
#Proceso FSFC: [tiempo_llegada, tiempo_ejecucion, id_proceso]

def fsfc(lista_proceso):
  tiempo_actual = 0
  gantt_final= []
  resultados_final = {}

  #ordenar por timepo de llegada
  lista_proceso.sort(key=lambda x: x[0])

  for proceso in lista_proceso:
    llegada = proceso[0]
    ejecucion = proceso[1]
    pid = proceso[2]

    if tiempo_actual < llegada:
      gantt.append("Idle")
      tiempo_actual = llegada

    inicio = tiempo_actual
    tiempo_actual += ejecucion
    final = tiempo_actual

    retorno = final - llegada
    espera = retorno - ejecucion

    resultados[pid] = [inicio,final, retorno, espera]
    gantt.append(pid)

procesos_ejemplo = [
    [0, 5, 'P1'],
    [2, 3, 'P2'],
    [4, 2, 'P3']
]



print("Diagrama de Gantt:")
print(gantt_final)

print("\nResulatdos")
for pid in resultados_final:
  print(pid, "Inicio:", resultados_final[pid][0],
        "Final:",resultados_final[pid][1],
        "Retorno:", resultados_final[pid][2],
        "Espera:", resultados_final[pid][3])



```
